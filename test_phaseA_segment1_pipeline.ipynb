```python
import cbgt as cbgt
from frontendhelpers import * 
from tracetype import *
import init_params as par 
import popconstruct_copy as popconstruct
import pdb
```

```python
pl = cbgt.Pipeline()
```

```python
#DEFAULTS - init_params.py
pl.celldefaults = pl[par.helper_cellparams]()
pl.popspecific = pl[par.helper_popspecific]()
pl.receptordefaults = pl[par.helper_receptor]()
pl.basestim = pl[par.helper_basestim]()
pl.dpmndefaults = pl[par.helper_dpmn]()
pl.d1defaults = pl[par.helper_d1]()
pl.d2defaults = pl[par.helper_d2]()
pl.actionchannels = pl[par.helper_actionchannels]()
pl.popdata = pl[popconstruct.helper_popconstruct](pl.actionchannels, pl.popspecific, pl.celldefaults, pl.receptordefaults, pl.basestim, pl.dpmndefaults, pl.d1defaults, pl.d2defaults)
pl.pathways = pl[popconstruct.helper_poppathways](pl.popdata)
#pl.connectivity_AMPA = pl[popconstruct.helper_connectivityAMPA](pl.popdata, pl.pathways)
#pl.connectivity_GABA, pl.meaneff_GABA = pl[popconstruct.helper_connectivityGABA](pl.popdata, pl.pathways)
#pl.connectivity_NMDA, pl.meaneff_NMDA = pl[popconstruct.helper_connectivityNMDA](pl.popdata, pl.pathways)
```

```python
#MODIFIERS - init_params.py 

def codeblock_modifycelldefaults(self):
    self.celldefaults = par.helper_cellparams(self.celldefaults, self.params)
    
def codeblock_modifypopspecific(self):
    self.popspecific = par.helper_popspecific(self.popspecific, self.pops)
    
def codeblock_modifyreceptordefaults(self):
    self.receptordefaults = par.helper_receptor(self.receptordefaults, self.receps)

def codeblock_modifybasestim(self):
    self.basestim_modified = par.modifier_basestim(self.basestim, self.base)
    
def codeblock_modifydpmndefaults(self):
    self.dpmndefaults_modified = par.modifier_dpmndefaults(self.dpmndefaults, self.dpmns)
    
def codeblock_modifyd1defaults(self):
    self.d1defaults_modified = par.modifier_d1defaults(self.d1defaults, self.d1)
    
def codeblock_modifyd2defaults(self):
    self.d2defaults_modified = par.modifier_d2defaults(self.d2defaults, self.d2)

def codeblock_modifyactionchannels(self): 
    self.actionchannels_modified = par.modifier_actionchannels(self.actionchannels, self.channels)
    
#POPDATA + PATHWAYS - popconstruct_copy.py

def codeblock_popconstruct_modified(self):
    self.popdata_modified = popconstruct.helper_popconstruct(self.actionchannels_modified, self.popspecific_modified, self.celldefaults_modified, self.receptordefaults_modified, self.basestim_modified, self.dpmndefaults_modified, self.d1defaults_modified, self.d2defaults_modified)
    
def codeblock_poppathways_modified(self): 
    self.pathways_modified = popconstruct.helper_poppathways(self.popdata_modified, self.newpathways) 

def codeblock_connAMPA_default(self): 
    self.connectivity_AMPA_default, self.meaneff_AMPA_defaults = popconstruct.helper_connectivityAMPA(self.popdata, self.pathways)
    
def codeblock_connAMPA_modified(self): 
    self.connectivity_AMPA_modified, self.meaneff_AMPA_modified = popconstruct.helper_connectivityAMPA(self.popdata_modified, self.pathways_modified)

def codeblock_connGABA_default(self): 
    self.connectivity_GABA_default, self.meaneff_GABA_default = popconstruct.helper_connectivityGABA(self.popdata, self.pathways) 
    
def codeblock_connGABA_modified(self): 
    self.connectivity_GABA_modified, self.meaneff_GABA_modified = popconstruct.helper_connectivityGABA(self.popdata_modified, self.pathways_modified) 

def codeblock_connNMDA_default(self): 
    self.connectivity_NMDA_default, self.meaneff_NMDA_default = popconstruct.helper_connectivityNMDA(self.popdata, self.pathways) 
    
def codeblock_connNMDA_modified(self): 
    self.connectivity_NMDA_modified, self.meaneff_NMDA_modified = popconstruct.helper_connectivityNMDA(self.popdata_modified, self.pathways_modified)    
```

```python
pl.add(codeblock_connAMPA_default)
pl.add(codeblock_connGABA_default)
pl.add(codeblock_connNMDA_default)
pl.add(codeblock_modifycelldefaults)
pl.add(codeblock_modifypopspecific)
pl.add(codeblock_modifyreceptordefaults)
pl.add(codeblock_modifybasestim)
pl.add(codeblock_modifydpmndefaults)
pl.add(codeblock_modifyd1defaults)
pl.add(codeblock_modifyd2defaults)
pl.add(codeblock_modifyactionchannels)
pl.add(codeblock_popconstruct_modified)
pl.add(codeblock_poppathways_modified)
pl.add(codeblock_connAMPA_modified)
pl.add(codeblock_connGABA_modified)
pl.add(codeblock_connNMDA_modified)
```

```python
environment = {
    'params': pd.DataFrame([[4, 1]], columns=['N', 'C']),
    'pops' : {'FSI': {'Taum': [60]}}, 
    'receps' : pd.DataFrame([[100, 100]], columns=['Tau_AMPA', 'Tau_GABA']), 
    'base' : {'FSI': {'FreqExt_AMPA': [100]}}, 
    'dpmns' : pd.DataFrame([[400, 100]], columns=['dpmn_wmax', 'dpmn_tauDOP']),
    'd1' : pd.DataFrame([[3, 50]], columns=['dpmn_type', 'dpmn_c']),
    'd2' : pd.DataFrame([[3, 50]], columns=['dpmn_type', 'dpmn_c']),
    'channels' : pd.DataFrame([[4],[5]], columns=['action']),
    'newpathways' : pd.DataFrame([['LIP', 'D1STR', 'AMPA', 'syn', 700, 0.027]], columns=['src', 'dest', 'receptor', 'type', 'con', 'eff'])
}
results = cbgt.ExecutionManager(cores=7).run(pl,environment) 
```
